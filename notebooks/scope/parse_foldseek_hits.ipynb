```python
import pandas as pd

# Function to count unique rows based on the first column in a .m8 file
def count_unique_rows_by_first_column(file_path):
    # Read the .m8 file into a pandas DataFrame
    df = pd.read_csv(file_path, sep='\t', header=None)

    # Get the number of unique values in the first column
    unique_rows_by_first_column = df.iloc[:, 0].nunique()

    return unique_rows_by_first_column, df

# Provide the path to your .m8 file
file_path = '/Users/seyonechithrananda/Documents/scope_aln_208.m8'

# Call the function and print the result
unique_count, df = count_unique_rows_by_first_column(file_path)
print(f'The number of unique rows based on the first column in the file is: {unique_count}')
```

```python
min_bit_score = -1
```

```python
import numpy as np
from tqdm import tqdm

D = np.zeros((len(test_df), len(lookup_df)), dtype=int)
I = np.zeros((len(test_df), len(lookup_df)), dtype=int)

for i in tqdm(range(len(df))):
    query = df.iloc[i]['query']
    target = df.iloc[i]['target']
    bit_score = df.iloc[i]['bit_score']
    query_idx = test_df[test_df['sid'] == query].index[0]
    target_idx = lookup_df[lookup_df['sid'] == target].index[0]
    D[query_idx, target_idx] = bit_score
    I[query_idx, target_idx] = target_idx

# iterate through all of D, and if an element is 0, set it to min_bit_score
# set I to j 
for i in range(len(test_df)):
    for j in range(len(lookup_df)):
        if D[i, j] == 0:
            D[i, j] = min_bit_score
            I[i, j] = j
```

```python
# return row's of D that have only -1's 
idx_to_drop = [i for i in range(len(test_df)) if np.all(D[i] == min_bit_score)]
idx_to_keep = [i for i in range(len(test_df)) if i not in idx_to_drop]

len(idx_to_drop) + len(idx_to_keep) == len(test_df)
```

```python
I 
```

```python
# Sort the rows of D to be in descending order of the bit scores, and rearrrange the rows of I accordingly in the same order
D_filtered = D[idx_to_keep]
#I_filtered = I[idx_to_keep]

I = np.argsort(-D_filtered, axis=1)
# Sort the matrix row-wise using the indices
sorted_D = np.take_along_axis(D_filtered, I, axis=1)
```

```python
def scope_hierarchical_loss(y_sccs, y_hat_sccs):
    """
    Find the common ancestor of two sets of SCCs (0 if family, 1 if superfamily, 2 if fold, 3 if class)
    """
    y_sccs, y_hat_sccs = y_sccs.split('.'), y_hat_sccs.split('.')
    first_non_matching_index = next((i for i, (x, y) in enumerate(zip(y_sccs, y_hat_sccs)) if x != y), len(y_sccs))

    loss = len(y_sccs) - first_non_matching_index # ex if the first mismatch is at idx 2 (0-indexed), that means that the second last label (superfamily) is wrong, which is a loss of two (wrong family & superfamily)
    exact = True if len(y_sccs) == first_non_matching_index else False

    return loss, exact
```

```python
near_ids = []

for i, idx in tqdm(enumerate(idx_to_keep)):
    test_id = test_df.loc[idx, 'sid']
    test_sccs = test_df.loc[idx, 'sccs']
    query_ids = [lookup_df.loc[j, 'sid'] for j in I[i]]
    exact_loss = [scope_hierarchical_loss(test_sccs, lookup_df.loc[j, 'sccs']) for j in I[i]]
    # grab the 2nd element in the tuple belonging to each element of exact_loss as mask_exact
    mask_exact = [x[1] for x in exact_loss]
    loss = [x[0] for x in exact_loss]
    
    # define mask_partial as 1 for any element of loss that is <=1 (tolerate retrieving homolog with diff family but same superfamily)
    mask_partial = [l <= 1 for l in loss]

    #mask_exact = [test_sccs == lookup_df.loc[lookup_idcs[j], 'sccs'] for j in I[i]]
    near_ids.append({
        'test_id': test_id,
        'query_ids': query_ids,
        #'meta_query': meta_query,
        'loss' : loss,
        'exact': mask_exact,
        'partial': mask_partial,
        'S_i': sorted_D[i],
        'I_i': I[i]
    })
```
